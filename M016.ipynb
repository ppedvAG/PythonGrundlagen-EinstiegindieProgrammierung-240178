### Shortcuts

Wenn wir auf der Zelle oben sind...
- Pfeil Hoch/Runter: Navigieren
- Strg+Shift+Hoch/Runter: Bewegt eine Zelle
- A/B: Erzeugen neue Zellen darüber oder darunter
- D+D: Zelle löschen
- C, V, X (ohne Strg): Copy, Paste und Cut
- Z und Shift+Z: Rückgängig, Wiederholen
- M und Y: Zwischen Markdown und Code umschalten
- Enter: Zelle betreten
- Shift+Enter, Strg+Enter: Zelle ausführen
- Escape: Zelle verlassen, ohne sie auszuführen
